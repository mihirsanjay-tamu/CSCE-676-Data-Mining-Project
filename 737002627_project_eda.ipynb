````python
# ============================================================================
# 5. VALUE COLUMN - Check numeric consistency
# ============================================================================
print("\n\n5. VALUE COLUMN - Numeric Consistency")
print("-" * 100)
print(f"Data type: {df[' Value'].dtype}")
print(f"Non-null values: {df[' Value'].notna().sum()}")
print(f"Null values: {df[' Value'].isna().sum()}")

# Check unique values
unique_values = df[' Value'].dropna().unique()
unique_values_sorted = sorted([str(v) for v in unique_values])
print(f"\nUnique Values ({len(unique_values_sorted)}):")
# Show a sample since there are many
print(f"  Sample: {unique_values_sorted[:20]}...")

# Try to clean currency values and convert to numeric
def clean_currency(val):
    if pd.isna(val):
        return np.nan
    val_str = str(val).replace('$', '').replace(',', '')
    try:
        return float(val_str)
    except:
        return np.nan

df[' Value_numeric'] = df[' Value'].apply(clean_currency)

# Check for non-numeric values after cleaning
non_numeric_mask = df[' Value'].notna() & (df[' Value_numeric'].isna())
non_numeric_count = non_numeric_mask.sum()

if non_numeric_count > 0:
    print(f"\n⚠️  ISSUE: {non_numeric_count} values could not be parsed as currency")
    print(f"Examples: {df[non_numeric_mask][' Value'].unique()[:10].tolist()}")
else:
    print("\n✓ All values successfully parsed as currency")

# Check for negative or zero values
invalid_values = df[(df[' Value_numeric'].notna()) & (df[' Value_numeric'] <= 0)]
if len(invalid_values) > 0:
    print(f"\n⚠️  ISSUE: {len(invalid_values)} rows with zero or negative values")
    print("Examples:")
    print(invalid_values[[' Round', ' Category', ' Value']].head())
else:
    print("✓ No zero or negative values found")

# Check for inconsistent formatting (with vs without commas)
values_without_comma = df[(df[' Value'].notna()) & (~df[' Value'].str.contains(',', na=False))][' Value'].unique()
if len(values_without_comma) > 0:
    print(f"\n⚠️  FORMATTING ISSUE: {len(df[(df[' Value'].notna()) & (~df[' Value'].str.contains(',', na=False))])} values lack comma separators")
    print(f"  Examples: {values_without_comma[:10].tolist()}")
else:
    print("✓ All values have consistent comma formatting")
````

**Now, let's check if there are any other inconsistencies in the Category column (same categories but different casing/punctuation etc.)**

````python
# ============================================================
# CHECK FOR CATEGORY INCONSISTENCIES (NO NORMALIZATION APPLIED)
# ============================================================

import re

unique_cats = df[' Category'].unique()

def key_for_check(cat):
    """Create comparison key only for checking inconsistencies"""
    cat = cat.upper()
    cat = re.sub(r'\s+', ' ', cat).strip()
    cat = re.sub(r'[.!?…]+$', '', cat)
    return cat

groups = {}

for cat in unique_cats:
    key = key_for_check(cat)
    groups.setdefault(key, []).append(cat)

# Keep only cases with multiple variants
duplicates = {k: v for k, v in groups.items() if len(v) > 1}

print(f"Potential inconsistent category groups: {len(duplicates)}")

if duplicates:
    print("\nExamples:")
    for i, (k, vals) in enumerate(list(duplicates.items())[:5], 1):
        print(f"\nExample {i}:")
        for v in vals:
            print(f"  '{v}'")
````

- As we can see , there are some more inconsistencies in the category column with regards to cases/spelling/spacing. Let's standardize these.

````python
# ============================================================
# STANDARDIZE CATEGORY COLUMN (FIX ALL FORMATTING INCONSISTENCIES)
# ============================================================

import re

def normalize_category(cat):
    if pd.isna(cat):
        return cat

    # 1. Standardize case
    cat = cat.upper()

    # 2. Remove extra spaces
    cat = re.sub(r'\s+', ' ', cat).strip()

    # 3. Remove trailing punctuation like ..., !, ?, etc.
    cat = re.sub(r'[.!?…]+$', '', cat).strip()

    # 4. Standardize quotes around letters ("I"s vs "I"S)
    cat = re.sub(r'"([A-Z])"S', r'"\1"s', cat)

    # 5. Normalize D.C. formatting
    cat = re.sub(r',?\s*D\.C\.', ' DC', cat)

    return cat

# Apply normalization
before = df[' Category'].nunique()
df[' Category'] = df[' Category'].apply(normalize_category)
after = df[' Category'].nunique()

print("Unique categories before:", before)
print("Unique categories after:", after)
print("Duplicates resolved:", before - after)
````

- **Now, let's explore which categories tend to have higher or lower prize values?**

````python
# ============================================================
# CATEGORY DIFFICULTY PROFILING (with clearer column names)
# ============================================================

# Aggregate stats per category
cat_difficulty = (
    df[df['Value_Numeric'].notna()]
    .groupby(' Category')['Value_Numeric']
    .agg(
        num_questions_in_category='count',
        avg_question_value='mean',
        median_question_value='median'
    )
    .sort_values('avg_question_value', ascending=False)
)

# Top hardest and easiest categories
hardest = cat_difficulty.head(10)
easiest = cat_difficulty.sort_values('avg_question_value').head(10)

print("\nHardest categories (highest average value):")
display(hardest)

print("\nEasiest categories (lowest average value):")
display(easiest)

# ============================================================
# Visualization
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Hardest categories
axes[0].barh(
    hardest.index[::-1],
    hardest['avg_question_value'][::-1]
)
axes[0].set_title("Highest Average Question Value")
axes[0].set_xlabel("Average Question Value ($)")
axes[0].set_ylabel("Category")
axes[0].grid(axis='x', alpha=0.3)

# Easiest categories
axes[1].barh(
    easiest.index[::-1],
    easiest['avg_question_value'][::-1]
)
axes[1].set_title("Lowest Average Question Value")
axes[1].set_xlabel("Average Question Value ($)")
axes[1].set_ylabel("Category")
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
````

- Assuming that questions having a higher "$" value are "more difficult", we see that the categories labeled “hardest” have higher average dollar values, but often contain very few questions. This assumption is not perfect, but pretty reasonable in my opinion.

- Many of the “easiest” categories appear only once and correspond to lowest prize values, typically early-round questions (since $100 is not possible in the double jeopardy round). An important note to make here is that, as the game evolved over time, there were differences in the minimum and maximum amounts a round could have per question.

- Question value (used as a proxy for difficulty) varies significantly across categories, though some categories with high average values contain very few questions, making difficulty estimation noisy.

Hypothesis:
- Manual category labels do not fully capture the semantic structure of questions, and questions across different categories may still share underlying themes.
- Topic and semantic patterns in Jeopardy questions evolve over time as show themes shift.
- Automated grouping and selection of questions can generate coherent Jeopardy-style boards without manual curation.

Potential RQs:
- How do themes and knowledge domains in Jeopardy questions evolve over time, and what latent topic structures emerge across decades of episodes?
- Can we automatically group Jeopardy questions into meaningful semantic clusters beyond manually assigned categories?
- Can we automatically generate coherent and difficulty-balanced Jeopardy-style question boards from historical question data?
